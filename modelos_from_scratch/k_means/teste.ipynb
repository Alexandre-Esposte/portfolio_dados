```python
from KMeans import KMeansScratch
import pandas as pd
from sklearn.datasets import make_blobs
import numpy as np
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
centros = 5
X, y = make_blobs(n_samples=500, centers=centros, n_features=100,cluster_std=0.2, random_state=42)
df = pd.DataFrame(X)
kmeansscratch = KMeansScratch(centros, rodadas=100)
kmeansscratch.fit(df)

kmeans = KMeans(centros,init='random')
kmeans.fit(df)
```

```python
import matplotlib.pyplot as plt

for cluster,rows in kmeansscratch.atribuicoes.items():
    df_aux = df.iloc[rows]
    plt.scatter(x=df_aux[0],y=df_aux[1],label=cluster)



for i,centroide in enumerate(kmeansscratch.centroids):
    plt.scatter(x = centroide[0],y=centroide[1],marker='x',linewidths=5,s=70,c='black')

#for i,centroide in enumerate(kmeansscratch.inicio):
#    plt.scatter(x = centroide[0],y=centroide[1],marker='.',linewidths=5,s=100,c='black')

plt.title(f'{kmeansscratch.inertia}')

plt.legend()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=df[0],y=df[1])
for i,centroide in enumerate(kmeans.cluster_centers_):
    plt.scatter(x = centroide[0],y=centroide[1],marker='x',linewidths=5,c='black',s=50,label=i)

plt.title(f'{kmeans.inertia_ }')

#plt.legend()
```
