```python
from KMeans_scratch import KMeansScratch
import pandas as pd
from sklearn.datasets import make_blobs
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
centros = 5
X, y = make_blobs(n_samples=500, centers=centros, n_features=1000,cluster_std=0.5, random_state=42)
df = pd.DataFrame(X)
kmeansscratch = KMeansScratch(centros, rodadas=500)
kmeansscratch.fit(df)

kmeans = KMeans(centros,init='random', n_init='auto')
kmeans.fit(df);
```

```python
fig , ax = plt.subplots(1,2, constrained_layout = True,figsize=(12,5))

for cluster,rows in kmeansscratch.atribuicoes.items():
    df_aux = df.iloc[rows]
    ax[0].scatter(x=df_aux[0],y=df_aux[1],label=cluster)

for i,centroide in enumerate(kmeansscratch.centroids):
    ax[0].scatter(x = centroide[0],y=centroide[1],marker='x',linewidths=5,s=70,c='black')


plt.scatter(x=df[0],y=df[1])
for i,centroide in enumerate(kmeans.cluster_centers_):
    ax[1].scatter(x = centroide[0],y=centroide[1],marker='x',linewidths=5,c='black',s=50,label=i)

#ax[0].set_title(f'Inertia :{kmeansscratch.inertia}')
ax[0].set_title(f"Scratch",size = 14)
ax[1].set_title(f"Sklearn",size = 14)

ax[0].legend();
```

```python
fig , ax = plt.subplots(1,2, constrained_layout = True,figsize=(12,5))

for cluster,rows in kmeansscratch.atribuicoes.items():
    df_aux = df.iloc[rows]
    ax[0].scatter(x=df_aux[0],y=df_aux[1],label=cluster)

for i,centroide in enumerate(kmeansscratch.centroids):
    ax[0].scatter(x = centroide[0],y=centroide[1],marker='x',linewidths=5,s=70,c='black')


plt.scatter(x=df[0],y=df[1])
for i,centroide in enumerate(kmeans.cluster_centers_):
    ax[1].scatter(x = centroide[0],y=centroide[1],marker='x',linewidths=5,c='black',s=50,label=i)

#ax[0].set_title(f'Inertia :{kmeansscratch.inertia}')
ax[0].set_title(f"Scratch\nInertia: {kmeansscratch.inertia:.2f}",size = 14)
ax[1].set_title(f"Sklearn\nInertia: {kmeans.inertia_:.2f}",size = 14)

ax[0].legend();
```
